# Simple PySpark notebook fixture
One read + one write, nothing fancy.

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('nb_simple').getOrCreate()
```

```python
orders = spark.read.format('parquet').load('s3://raw/orders/')
orders.write.format('delta').mode('overwrite').saveAsTable('prod.mart.orders_nb')
```
