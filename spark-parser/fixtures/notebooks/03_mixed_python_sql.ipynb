```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('nb_mixed').getOrCreate()
```

```python
raw = spark.read.format('parquet').load('s3://raw/orders/')
raw.createOrReplaceTempView('raw_orders')
```

```python
CREATE OR REPLACE TABLE prod.mart.orders_nb_sql AS
SELECT order_id, customer_id, amount
FROM raw_orders
WHERE amount > 0
```

```python
INSERT INTO prod.mart.orders_summary_nb
SELECT customer_id, SUM(amount) AS total
FROM prod.mart.orders_nb_sql
GROUP BY customer_id
```
